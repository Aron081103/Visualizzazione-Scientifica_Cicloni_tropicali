```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap

df = pd.read_csv('../data/casi_cicloni.csv')
df['Data'] = pd.to_datetime(df['Data'])

def plot_ciclone(df, nome, titolo):
    sotto_df = df[df['Nome'] == nome]
    date = sotto_df['Data'].dt.strftime('%d %b')
    categorie = sotto_df['Categoria']

    cmap = get_cmap("Blues")  
    colors = [cmap(cat / 5) for cat in categorie]  

    plt.figure(figsize=(9, 5))
    plt.bar(date, categorie, color=colors, edgecolor="black")
    plt.title(titolo, fontsize=16)
    plt.xlabel('Data', fontsize=12)
    plt.ylabel('Categoria Saffir–Simpson', fontsize=12)
    plt.ylim(0, 5)
    plt.yticks(range(6))
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_ciclone(df, 'Katrina', 'Hurricane Katrina – Categoria Saffir-Simpson (2005)')
plot_ciclone(df, 'Doksuri', 'Tifone Doksuri – Categoria Saffir-Simpson (2023)')
plot_ciclone(df, 'Nargis', 'Ciclone Nargis – Categoria Saffir-Simpson (2008)')
```
