```python
import matplotlib.pyplot as plt


months = ['Gen', 'Feb', 'Mar', 'Apr', 'Mag', 'Giu', 
          'Lug', 'Ago', 'Set', 'Ott', 'Nov', 'Dic']
month_indices = list(range(12))

fig, ax = plt.subplots(figsize=(12, 4))
ax.set_xlim(0, 12)
ax.set_ylim(-0.5, 3)

bar_height = 0.6

y_uragani = 2
y_cicloni = 1
y_tifoni = 0

ax.broken_barh([(5, 6)], (y_uragani, bar_height), facecolors='#1f77b4', edgecolor='black')
ax.broken_barh([(10, 2), (0, 4)], (y_cicloni, bar_height), facecolors='#4a90e2', edgecolor='black')
ax.broken_barh([(4, 6)], (y_tifoni, bar_height), facecolors='#0d47a1', edgecolor='black')

ax.set_yticks([y_tifoni + bar_height/2, y_cicloni + bar_height/2, y_uragani + bar_height/2])
ax.set_yticklabels(['Tifoni', 'Cicloni', 'Uragani'], fontsize=12)

ax.set_xticks(month_indices)
ax.set_xticklabels(months, fontsize=11)

ax.set_title('Stagionalit√† dei principali cicloni tropicali', fontsize=16, weight='bold')

ax.grid(True, axis='x', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```
