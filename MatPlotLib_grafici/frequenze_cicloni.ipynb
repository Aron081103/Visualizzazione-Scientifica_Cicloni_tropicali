```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("..\\data\\frequenze_cicloni.csv")

anni = df["Anno"]
tropical_storms = df["Tropical_Storms"]
hurricanes = df["Hurricanes"]
major_hurricanes = df["Major_Hurricanes"]  


color_tropical = '#6baed6'  
color_hurricanes = '#2171b5' 
color_major = '#08306b'  

plt.figure(figsize=(14, 7))
plt.plot(anni, tropical_storms, 'o-', color=color_tropical, label='Cicloni tropicali')
plt.plot(anni, hurricanes, 'o-', color=color_hurricanes, label='Uragani/Cicloni/Tifoni')
plt.plot(anni, major_hurricanes, 'o-', color=color_major, label='Cicloni tropicali maggiori')


plt.text(anni.iloc[-1] + 0.2, tropical_storms.iloc[-1], f"{int(round(tropical_storms.iloc[-1]))}", color=color_tropical, fontsize=12)
plt.text(anni.iloc[-1] + 0.2, hurricanes.iloc[-1], f"{int(round(hurricanes.iloc[-1]))}", color=color_hurricanes, fontsize=12)
plt.text(anni.iloc[-1] + 0.2, major_hurricanes.iloc[-1], f"{int(round(major_hurricanes.iloc[-1]))}", color=color_major, fontsize=12)

plt.title("Frequenza Globale dei Cicloni Tropicali ogni 6 mesi", fontsize=24)
plt.xlabel("Anno", fontsize=18)
plt.ylabel("Numero di cicloni tropicali", fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlim(1970, 2025)
plt.ylim(0, 120)
plt.legend()
plt.tight_layout()
plt.show()
```
