```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../data/vittime_cicloni.csv')
plt.figure(figsize=(12,6))
plt.plot(df['Anno'], df['Vittime'], marker='o', linestyle='-', color='teal', linewidth=2, markersize=6)
plt.yscale('log')
plt.xlabel('Anno', fontsize=18)
plt.ylabel('Morti da cicloni tropicali (scala log)', fontsize=18)
plt.title('Vittime da cicloni tropicali: anni selezionati (2004‑2024)', fontsize=24)
plt.xticks(df['Anno'], rotation=45, fontsize=15)
plt.yticks(fontsize=15)
plt.grid(which='both', axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
