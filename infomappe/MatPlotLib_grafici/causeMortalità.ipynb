```python
import matplotlib.pyplot as plt

cause = ['Marea', 'Pioggia', 'Wind', 'Altro', 'Onde']  
percentuali = [47, 27, 11, 3, 12]  
colori = ['#c6dbef', '#9ecae1', '#6baed6', '#3182bd', "#3872ad"]  

plt.figure(figsize=(8,8))
plt.pie(percentuali, labels=cause, colors=colori, startangle=140, 
        autopct='%1.0f%%', textprops={'fontsize': 14})
plt.title('Cause di mortalità da cicloni tropicali dal 1983 a 2015', fontsize=17, pad=20)
plt.axis('equal')
plt.show()
```
