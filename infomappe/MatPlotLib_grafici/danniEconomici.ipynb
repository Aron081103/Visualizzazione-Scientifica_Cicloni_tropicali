```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../data/ciclone_danni.csv')


colori = {
    'Uragano': '#08306b', 
    'Tifone': '#2171b5',   
    'Ciclone': '#6baed6'  
}

bar_colors = [colori[tipo] for tipo in df['Tipo']]

plt.figure(figsize=(12,7))
bars = plt.bar(df['Nome'], df['Danni'], color=bar_colors)
plt.ylabel('Danni economici (miliardi USD)', fontsize=18)
plt.title('Danni economici dei cicloni tropicali degli ultimi anni', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar, val in zip(bars, df['Danni']):
    plt.text(bar.get_x() + bar.get_width()/2, val + 5, f'{val}', ha='center', fontsize=9)
handles = [plt.Rectangle((0,0),1,1, color=colori[tipo]) for tipo in colori]
plt.legend(handles, colori.keys(), title='Tipo di evento')
plt.tight_layout()
plt.show()
```
