```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches


lat = np.array([15, 16, 17, 18, 19, 20])
lon = np.array([-60, -61, -62, -63, -64, -65])


radii = np.array([0.5, 1, 1.5, 2, 3, 4])

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(lon, lat, color='blue', marker='o', label='Traiettoria prevista')

for i in range(len(lat)):
    circle = plt.Circle((lon[i], lat[i]), radii[i], color='#156082', alpha=0.2)
    ax.add_artist(circle)

np.random.seed(0)
num_alt = 7  
for n in range(num_alt):
    alt_lat = lat + np.random.uniform(-0.5, 0.5, size=len(lat)) * (radii / radii[0])
    alt_lon = lon + np.random.uniform(-0.5, 0.5, size=len(lon)) * (radii / radii[0])
    ax.plot(alt_lon, alt_lat, color='blue', linestyle='--', alpha=0.5)

uncertainty_patch = mpatches.Patch(color='#156082', alpha=0.2, label="Cono d'incertezza")
ax.legend(handles=[plt.Line2D([], [], color='blue', marker='o', label='Traiettoria prevista'),
                   uncertainty_patch])
ax.set_xlabel("Longitudine")
ax.set_ylabel("Latitudine")
ax.set_title("Cono d'incertezza di un ciclone tropicale")
ax.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
