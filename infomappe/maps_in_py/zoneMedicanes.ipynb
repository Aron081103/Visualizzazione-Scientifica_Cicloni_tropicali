```python
import pandas as pd
import folium
from branca.element import Template, MacroElement

df_zone = pd.read_csv('..\\data\\zone_medicanes.csv')

mappa = folium.Map(location=[36, 18], zoom_start=5, tiles='CartoDB positron')

def add_zone(bounds, nome):
    folium.Rectangle(
        bounds=bounds,
        color='#156082',
        fill=True,
        fill_opacity=0.2,
        fill_color='#156082',
        tooltip=nome
    ).add_to(mappa)

for idx, row in df_zone.iterrows():
    bounds = [[row['Lat_min'], row['Lon_min']], [row['Lat_max'], row['Lon_max']]]
    add_zone(bounds, row['Nome zona'])

template = """
{% macro html(this, kwargs) %}
<style>
    #map_title {
        position: fixed;
        top: 10px;
        left: 50%;
        transform: translateX(-50%);
        z-index: 9999;
        background-color: white;
        padding: 5px 15px;
        font-size: 18px;
        font-weight: bold;
        border-radius: 5px;
        box-shadow: 2px 2px 5px rgba(0,0,0,0.3);
    }
    .legend {
        position: fixed;
        bottom: 30px;
        left: 10px;
        z-index: 9999;
        background-color: white;
        padding: 10px;
        border-radius: 5px;
        box-shadow: 2px 2px 5px rgba(0,0,0,0.3);
        font-size: 14px;
    }
    .legend-title {
        font-weight: bold;
        margin-bottom: 5px;
    }
    .legend-item {
        display: flex;
        align-items: center;
        margin-bottom: 3px;
    }
    .legend-color {
        width: 20px;
        height: 20px;
        margin-right: 5px;
        background:#156082;
    }
</style>

<div id="map_title">Zone Medicanes nel Mediterraneo</div>
<div class="legend">
  <div class="legend-title">Legenda</div>
  <div class="legend-item">
    <div class="legend-color"></div>
    <div>Zone di interesse</div>
  </div>
</div>

{% endmacro %}
"""

macro = MacroElement()
macro._template = Template(template)
mappa.get_root().add_child(macro)

mappa.save("medicanes_zone_blu.html")
```
