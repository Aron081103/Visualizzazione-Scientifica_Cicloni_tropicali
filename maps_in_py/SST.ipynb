```python
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

ds = xr.open_dataset("..\\data\\sst.mnmean.nc")
sst = ds['sst'].sel(time=slice('2023-01', '2023-12')).mean(dim='time')
plt.figure(figsize=(14, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
sst.plot(
    ax=ax,
    transform=ccrs.PlateCarree(),
    cmap='turbo',
    cbar_kwargs={'label': 'Temperatura (°C)'},
    levels=range(0, 33, 1)
)
ax.coastlines()
ax.set_title('Temperatura Superficiale del Mare - Media Annuale 2023', fontsize=16)
ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.5)
plt.tight_layout()
plt.show()
```
